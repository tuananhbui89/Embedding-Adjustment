```python
import os

def generate_ti_scripts(cs101_settings, prefix):
    os.makedirs('scripts', exist_ok=True)
    for name, settings in cs101_settings.items():
        
        script_path = f"scripts/run_ti/run_{prefix}_{name}.sh"
        os.makedirs('scripts/run_ti', exist_ok=True)

        with open(script_path, 'w') as f:
            f.write('#!/bin/bash\n\n')
            prompt_file_name = settings["prompt_file"].split("/")[-1].split(".")[0]
                        
            # # Training command
            f.write(f"CUDA_VISIBLE_DEVICES=0 accelerate launch --num_processes=1 textual_inversion.py \\\n")
            f.write(f"  --pretrained_model_name_or_path=\"stable-diffusion-v1-5/stable-diffusion-v1-5\" \\\n")
            f.write(f"  --train_data_dir=\"{settings['data_dir']}\" \\\n")
            f.write(f"  --learnable_property=\"object\" \\\n")
            f.write(f"  --placeholder_token=\"{settings['placeholder_token']}\" --initializer_token=\"{settings['initializer_token']}\" \\\n")
            f.write(f"  --resolution=512 \\\n")
            f.write(f"  --train_batch_size=1 \\\n")
            f.write(f"  --gradient_accumulation_steps=1 \\\n")
            f.write(f"  --max_train_steps=3000 \\\n")
            f.write(f"  --learning_rate=5.0e-04 --scale_lr \\\n")
            f.write(f"  --lr_scheduler=\"constant\" \\\n")
            f.write(f"  --lr_warmup_steps=0 \\\n")
            f.write(f"  --output_dir=\"outputs/{prefix}_{name}\" \\\n")
            f.write(f"  --validation_steps=100 \\\n")
            f.write(f"  --save_steps=100 \\\n")
            f.write(f"  --checkpointing_steps=10000 \\\n")
            f.write(f"  --mixed_precision=\"no\" \\\n")
            f.write(f"  --validation_prompt=\"a photo of a {settings['placeholder_token']} wearing glasses and writing on a red notebook\"\n\n")

            # # Generate batch of images for evaluation at step 1000
            f.write(f"CUDA_VISIBLE_DEVICES=0 python generate_textual_inversion.py \\\n")
            f.write(f"    --method=\"distill_ti\" \\\n")
            f.write(f"    --model_path=\"stable-diffusion-v1-5/stable-diffusion-v1-5\" \\\n")
            f.write(f"    --concept_name=\"{settings['placeholder_token']}\" \\\n")
            f.write(f"    --output_dir=\"evaluation_massive/{prefix}_{name}_steps_1000\" \\\n")
            f.write(f"    --guidance_scale=7.0 \\\n")
            f.write(f"    --embedding_path=\"outputs/{prefix}_{name}/learned_embeds-steps-1000.bin\" \\\n")
            f.write(f"    --max_train_steps=1000 \\\n")
            f.write(f"    --start_steps=0 \\\n")
            f.write(f"    --num_images=50 \\\n")
            f.write(f"    --prompt_file=\"{settings['prompt_file']}\"\n\n")
            
            # Run CLIP similarity evaluation with different custom prompts
            f.write("for custom_prompt in 'objectA' 'objectB' 'full_prompt'; do\n")
            f.write(f"    CUDA_VISIBLE_DEVICES=0 python investigate_clip_sim_v2.py \\\n")
            f.write(f"        --images_folder=\"evaluation_massive/{prefix}_{name}_steps_1000/{settings['placeholder_token']}/{prompt_file_name}\" \\\n")
            f.write(f"        --prompt_file=\"{settings['prompt_file']}\" \\\n")
            f.write(f"        --num_images=50 \\\n")
            f.write(f"        --use_custom_prompt=$custom_prompt \\\n")
            f.write(f"        --info=\"use_custom_prompt_${{custom_prompt}}\" \\\n")
            f.write(f"        --sub_folder=\"None\" \\\n")
            f.write(f"        --output_dir=\"semantic_drift/{prefix}_{name}_steps_1000/1000\"\n")
            f.write("done\n\n")
            
            # # Run CLIP image similarity evaluation
            f.write(f"CUDA_VISIBLE_DEVICES=0 python investigate_clip_sim_image.py \\\n")
            f.write(f"    --images_folder=\"evaluation_massive/{prefix}_{name}_steps_1000/{settings['placeholder_token']}/{prompt_file_name}\" \\\n")
            f.write(f"    --prompt_file=\"{settings['prompt_file']}\" \\\n")
            f.write(f"    --num_images=50 \\\n")
            f.write(f"    --anchor_image_path=\"{settings['anchor_image_path']}\" \\\n")
            f.write(f"    --info=\"t01\" \\\n")
            f.write(f"    --sub_folder=\"None\" \\\n")
            f.write(f"    --output_dir=\"semantic_drift/{prefix}_{name}_steps_1000/1000\"\n")
            f.write("\n\n")


            # # Run DINO image similarity evaluation
            f.write(f"CUDA_VISIBLE_DEVICES=0 python investigate_dino_sim_image.py \\\n")
            f.write(f"    --images_folder=\"evaluation_massive/{prefix}_{name}_steps_1000/{settings['placeholder_token']}/{prompt_file_name}\" \\\n")
            f.write(f"    --prompt_file=\"{settings['prompt_file']}\" \\\n")
            f.write(f"    --num_images=50 \\\n")
            f.write(f"    --anchor_image_path=\"{settings['anchor_image_path']}\" \\\n")
            f.write(f"    --info=\"t01\" \\\n")
            f.write(f"    --sub_folder=\"None\" \\\n")
            f.write(f"    --output_dir=\"semantic_drift/{prefix}_{name}_steps_1000/1000\"\n")
            f.write("\n\n")



        os.chmod(script_path, 0o755)
        print(f"{script_path}")

# Run the function with your experiment settings
generate_ti_scripts(cs101_settings, 'cs101_ti')
generate_ti_scripts(celebA_settings, 'celebA_ti')

```

```python
import os

def generate_ti_adjustment_scripts(cs101_settings, prefix):
    os.makedirs('scripts', exist_ok=True)
    for name, settings in cs101_settings.items():
        
        script_path = f"scripts/run_ti_adjustment/run_{prefix}_{name}.sh"
        os.makedirs('scripts/run_ti_adjustment', exist_ok=True)

        with open(script_path, 'w') as f:
            f.write('#!/bin/bash\n\n')
            prompt_file_name = settings["prompt_file"].split("/")[-1].split(".")[0]

            # Generate images with different alpha and rho values
            f.write("for alpha in 0.5; do\n")
            f.write("    for rho in 0.2; do\n")
            f.write(f"        CUDA_VISIBLE_DEVICES=0 python generate_ti_adjust_embedding_v3.py \\\n")
            f.write(f"            --method=\"standard\" \\\n")
            f.write(f"            --model_path=\"stable-diffusion-v1-5/stable-diffusion-v1-5\" \\\n")
            f.write(f"            --concept_name=\"{settings['placeholder_token']}\" \\\n")
            f.write(f"            --output_dir=\"evaluation_massive/{prefix}_{name}_adjust_embedding_v3/rho${{rho}}_alpha${{alpha}}/1000\" \\\n")
            f.write(f"            --guidance_scale=7.0 \\\n")
            f.write(f"            --embedding_path=\"outputs/{prefix}_{name}/learned_embeds-steps-1000.bin\" \\\n")
            f.write(f"            --max_train_steps=3000 \\\n")
            f.write(f"            --num_images=50 \\\n")
            f.write(f"            --start_prompt_index=0 \\\n")
            f.write(f"            --prompt_file=\"{settings['prompt_file']}\" \\\n")
            f.write(f"            --target_word=\"{settings['target_word']}\" \\\n")
            f.write(f"            --rho=$rho \\\n")
            f.write(f"            --alpha=$alpha\n")
            f.write("    done\n")
            f.write("done\n\n")
            
            # Run CLIP similarity evaluation with different custom prompts
            f.write("for alpha in 0.5; do\n")
            f.write("    for rho in 0.2; do\n")
            f.write("        for custom_prompt in 'objectA' 'objectB' 'full_prompt'; do\n")
            f.write(f"                CUDA_VISIBLE_DEVICES=0 python investigate_clip_sim_v2.py \\\n")
            f.write(f"                --images_folder=\"evaluation_massive/{prefix}_{name}_adjust_embedding_v3/rho${{rho}}_alpha0.5/1000/{settings['placeholder_token']}/{prompt_file_name}\" \\\n")
            f.write(f"                --prompt_file=\"{settings['prompt_file']}\" \\\n")
            f.write(f"                --num_images=50 \\\n")
            f.write(f"                --use_custom_prompt=$custom_prompt \\\n")
            f.write(f"                --info=\"use_custom_prompt_${{custom_prompt}}\" \\\n")
            f.write(f"                --sub_folder=\"None\" \\\n")
            f.write(f"                --output_dir=\"semantic_drift/{prefix}_{name}_adjust_embedding_v3/rho${{rho}}_alpha${{alpha}}/1000\"\n")
            f.write("        done\n")
            f.write("    done\n")
            f.write("done\n\n")
            
            # # Run CLIP image similarity evaluation
            f.write("for alpha in 0.5; do\n")
            f.write("    for rho in 0.2; do\n")
            f.write(f"        CUDA_VISIBLE_DEVICES=0 python investigate_clip_sim_image.py \\\n")
            f.write(f"            --images_folder=\"evaluation_massive/{prefix}_{name}_adjust_embedding_v3/rho${{rho}}_alpha0.5/1000/{settings['placeholder_token']}/{prompt_file_name}\" \\\n")
            f.write(f"            --prompt_file=\"{settings['prompt_file']}\" \\\n")
            f.write(f"            --num_images=50 \\\n")
            f.write(f"            --anchor_image_path=\"{settings['anchor_image_path']}\" \\\n")
            f.write(f"            --info=\"t01\" \\\n")
            f.write(f"            --sub_folder=\"None\" \\\n")
            f.write(f"            --output_dir=\"semantic_drift/{prefix}_{name}_adjust_embedding_v3/rho${{rho}}_alpha${{alpha}}/1000\"\n")
            f.write("    done\n")
            f.write("done\n")

            # Run DINO image similarity evaluation
            f.write("for alpha in 0.5; do\n")
            f.write("    for rho in 0.2; do\n")
            f.write(f"        CUDA_VISIBLE_DEVICES=0 python investigate_dino_sim_image.py \\\n")
            f.write(f"            --images_folder=\"evaluation_massive/{prefix}_{name}_adjust_embedding_v3/rho${{rho}}_alpha0.5/1000/{settings['placeholder_token']}/{prompt_file_name}\" \\\n")
            f.write(f"            --prompt_file=\"{settings['prompt_file']}\" \\\n")
            f.write(f"            --num_images=50 \\\n")
            f.write(f"            --anchor_image_path=\"{settings['anchor_image_path']}\" \\\n")
            f.write(f"            --info=\"t01\" \\\n")
            f.write(f"            --sub_folder=\"None\" \\\n")
            f.write(f"            --output_dir=\"semantic_drift/{prefix}_{name}_adjust_embedding_v3/rho${{rho}}_alpha${{alpha}}/1000\"\n")
            f.write("    done\n")
            f.write("done\n")

        os.chmod(script_path, 0o755)
        print(f"{script_path}")

# Run the function with your experiment settings
generate_ti_adjustment_scripts(cs101_settings, 'cs101_ti')
generate_ti_adjustment_scripts(celebA_settings, 'celebA_ti')
```
